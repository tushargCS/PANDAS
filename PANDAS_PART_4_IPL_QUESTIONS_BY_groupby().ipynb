```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("deliveries.csv")
df
```

```python
df.shape
```

```python
#find the Top 10 batsman in term of runs
df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
#find the name of the batsman with max no of six
df[df['batsman_runs']==6].groupby('batsman')['batsman'].count().sort_values(ascending=False).head(10)
```

```python
df['batsman_runs']==6
```

```python
mask = df[df['over'] > 15]
```

```python
mask2 = mask[(mask['batsman_runs']==4) | (mask['batsman_runs']==6)]
```

```python
mask2.groupby('batsman').count().sort_values('batsman_runs',ascending=False)
```

```python
mask2.groupby('batsman')['batsman'].count().sort_values(ascending=False).head()
```

```python
#find V kohli's record against all teams
x = df[df['batsman']=='V Kohli'].groupby('bowling_team').sum().reset_index().sort_values('batsman_runs',ascending=False)
x.head()
```

```python
#function for highest score of any batsman in ipl
def high(p):
    x = df[df['batsman']==p]
    return x.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
high('V Kohli')
```
