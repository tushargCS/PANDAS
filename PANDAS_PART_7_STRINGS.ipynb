# split()

```python
tit['Name'].str.split(',').head()
```

```python
tit['last_name'] = tit['Name'].str.split(',').str.get(0).head()
```

```python
tit.head()
```

```python
tit['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1)
```

```python
# Attributes of split()--> n=1,expand=True
tit['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
```

```python
tit[['title','first_name']]=tit['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
```

```python
tit.head()
```

```python
tit['title'].value_counts()
```

# replace

```python
tit['title'] = tit['title'].str.replace('Ms.','Miss.')
tit['title'] = tit['title'].str.replace('Mlle.','Miss.')
```

```python
tit['title'].value_counts()
```

# filtering

# .str.startswith

```python
# startswith/endswith
tit[tit['first_name'].str.startswith('A')].head()
```

# .str.endswith

```python
tit[tit['first_name'].str.endswith('A')]
```

# .str.isdigit()

```python
tit[tit['first_name'].str.isdigit()]
```

# Apply regex

```python
tit[tit['first_name'].str.contains('john',case=False)].head()
```

```python
# find lastname ends eith and starts with vowel
tit[tit['first_name'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```

```python
# not vowel
tit[tit['first_name'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```
