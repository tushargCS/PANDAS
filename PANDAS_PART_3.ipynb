# groupby()

generally we use groupby() on categorical cols

```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv(r"imdb-top-1000.csv")
df.head()
```

```python
genres = df.groupby('Genre')
```

```python
#apply builtin aggregate functions on groupby objects
genres.sum()
```

```python
genres.min()
```

```python
genres.mean()
```

```python
df.head()
```

```python
#find the top 3 genres by total earning
df.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
df.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
#find the genre with highest avg IMDB rating
df.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
#find director with most popularity on the basis of No_of_votes
df.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
#no of movies done by each actor
df.groupby('Star1')['Series_Title'].count().sort_values()
```

```python
df['Star1'].value_counts()
```

# agg method

```python
df.groupby('Genre').agg({
    'Runtime':'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum'
})
```

```python
#passing list
df.groupby('Genre').agg(['min','max','mean'])
```

```python
# adding both the sentex
df.groupby('Genre').agg(
{
    'Runtime':['min','max'],
    'IMDB_Rating':'mean'
}
)
```

# looping on groups

```python
for g,d in df.groupby('Genre'):
    print(type(g),type(d))
```

```python
for g,d in df.groupby('Genre'):
    print(d)
```

```python
#find highest rated movies of each genre
mo = pd.DataFrame(df,columns=df.columns)
for g,d in df.groupby('Genre'):
    mo = mo.append(d[d['IMDB_Rating']==d['IMDB_Rating'].max()])
```

```python
mo
```

```python
# split(apply) combine
df.groupby('Genre').apply(min)
```

```python
#find no. of movies starting with A for each group
def find(g):
      return  (g['Series_Title'].str[0]=='A').sum()
df.groupby('Genre').apply(find)
```

```python
df[df['Series_Title'].str[0]=='A'].groupby('Genre').size()
```

```python
#find ranking of each movie in the group according to IMDB score
def r(g):
    g['rank']=g['IMDB_Rating'].rank(ascending=False)
    return (g)
df.groupby('Genre').apply(r)
```

#Normalized Rating
formula:=-->  (x-xmin)/(xmax-xmin)

```python
#find normalized IMDB rating group-wise
def nor(g):
    g['Max_rating'] = (g['IMDB_Rating']-g['IMDB_Rating'].min())/( g['IMDB_Rating'].max()-g['IMDB_Rating'].min() )
    return g
```

```python
df.groupby('Genre').apply(nor).head()
```

# groupby on multiple column
