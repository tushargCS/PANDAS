```python
#write a function that can return the track record of 2 teams againts each other
def track(t1,t2):
    m1 = (df2['Team1']==t1)&(df2['Team2']==t2)
    m2 = (df2['Team1']==t2)&(df2['Team2']==t1)
    x = df2[m1|m2]
    print(t1," Win ",x.shape[0]-(x[x['WinningTeam']==t2].shape[0]))
    print(t2," Win ",x.shape[0]-(x[x['WinningTeam']==t1].shape[0]))
track('Royal Challengers Bangalore','Chennai Super Kings')   
```

```python
point_df =pd.DataFrame(
 {
     '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
     '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
 }
) 
point_df
```

```python
def euc(row):
    pt_a = row['1st point']
    pt_b = row['2nd point']
    return (pt_a[0]-pt_b[0])**2+(pt_a[1]-pt_b[1])**2
```

```python
x = point_df.apply(euc,axis=1)
```

```python
point_df['euc_dist']=x
point_df
```
